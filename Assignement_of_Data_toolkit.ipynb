6. What are the differences between NumPy arrays and Python lists?
* umPy arrays and Python lists are both data structures used to store collections of elements, but they have key differences:

-Performance: NumPy arrays are more efficient for numerical computations and large datasets due to vectorized operations and optimized memory management.

8. What does the term “vectorized operation” mean in NumPy?
* In NumPy, "vectorized operation" refers to performing operations on entire arrays at once, rather than iterating over individual elements. This approach:
-Speeds up computations: Operations are applied element-wise without explicit loops.
-Improves efficiency: Optimized C code underlies vectorized operations.
-Simplifies code: Vectorized operations often result in more concise and readable code.

3. Create a scatter plot using Matplotlib?

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
x1 = np.random.randn(50)
y1 = np.random.randn(50)
x2 = np.random.randn(50) + 2
y2 = np.random.randn(50) + 2

plt.scatter(x1, y1, label='Group 1', color='blue')
plt.scatter(x2, y2, label='Group 2', color='green')
plt.title('Scatter Plot with Labels and Legend')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
```

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a sample DataFrame
np.random.seed(0)
data = np.random.randn(100, 5)
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
```

6.Create a DataFrame and add a new column based on an existing column?

8.Create a line plot with multiple lines using Matplotlib?

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)

plt.plot(x, y1, label='Sine', linestyle='-', color='blue')
plt.plot(x, y2, label='Cosine', linestyle='--', color='red')
plt.plot(x, y3, label='Sine 2x', linestyle=':', color='green')

plt.title('Customized Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show
()
```

10. Create a histogram using Seaborn to visualize a distribution?


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
data = np.random.randint(1, 7, 1000)

plt.figure(figsize=(8, 6))
sns.histplot(data, bins=6, discrete=True, kde=False)

plt.title('Histogram with Discrete Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

13. Create a 3D scatter plot using Plotly?

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

df = pd.DataFrame({
    'x': x,
    'y': y,
    'z': z
})
fig = px.scatter_3d(df, x='x', y='y', z='z')

fig.update_layout(title='3D Scatter Plot')

# Display the plot
fig.show()
```
